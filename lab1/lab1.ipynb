```python
import numpy as np
import matplotlib.pyplot as plt
import math
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
X_START = -math.pi
X_END = math.pi
Y_START = -math.pi
Y_END = math.pi
Z_START = -3
Z_END = 3
```

```python
X = np.arange(X_START, X_END, 0.25)
Y = np.arange(Y_START, Y_END, 0.25)
X, Y = np.meshgrid(X, Y)
Z = X * np.sin(Y)
```

```python
fix, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X,Y,Z)
ax.set_zlim()
plt.title('3D график функции')
plt.show(Z_START, Z_END)
```

```python
TERMS = [
    [
        'очень малая',
        'малая',
        'средняя',
        'большая',
        'очень большая'
    ],
    [
        'сильно малая',
        'очень малая',
        'малая',
        'средняя',
        'большая',
        'очень большая',
        'сильно большая'
    ],
    [
        'сильно малая',
        'очень малая',
        'малая',
        'ниже среднего',
        'средняя',
        'выше среднего',
        'большая',
        'очень большая',
        'сильно большая'
    ],
    [
        'сильно малая',
        'очень малая',
        'малая',
        'выше малой',
        'ниже среднего',
        'средняя',
        'выше среднего',
        'ниже большой',
        'большая',
        'очень большая',
        'сильно большая'
    ]
]
```

```python
def data(start, end):
    for terms in TERMS:
        termsNum = len(terms)
        print("Для {} термов:".format(termsNum))

        step = (abs(start) + abs(end)) / termsNum
        for i in range(termsNum):
            print("{} на отрезке от {:.2f} до {:.2f}".format(terms[i], start + step * i, start + step * (i + 1)))
```

```python
print("Данные для функции принадлежности X:")
data(X_START, X_END)
```

```python
print("Данные для функции принадлежности Y:")
data(Y_START, Y_END)
```

```python
print("Данные для функции принадлежности Z:")
data(Z_START, Z_END)
```

```python
def ranges(start, end, variable):
    ranges = {}
    numTerm = 0

    for terms in TERMS:
        termsNum = len(terms)
        ranges[numTerm] = {}
        step = (abs(start) + abs(end)) / termsNum

        label = "Функция принадлежности {} при {} термах".format(variable, termsNum)
        graph = ctrl.Antecedent(np.arange(start, end + step, step), label)

        for i in np.arange(0, termsNum, 1):
            triangle = []
            if i == 0:
                triangle = [start, start, start + step]
                ranges[numTerm][i] = {}
                ranges[numTerm][i]['start'] = triangle[1]
                ranges[numTerm][i]['end'] = triangle[2] - (step / 2)
            else:
                triangle = [
                    (i - 1) * step - abs(start),
                    i * step - abs(start),
                    (i + 1) * step - abs(start)
                ]
                if termsNum - 1 == i:
                    ranges[numTerm][i] = {}
                    ranges[numTerm][i]['start'] = triangle[0] + (step / 2)
                    ranges[numTerm][i]['end'] = triangle[1]
                else:
                    ranges[numTerm][i] = {}
                    ranges[numTerm][i]['start'] = triangle[0] + (step / 2)
                    ranges[numTerm][i]['end'] = triangle[2] - (step / 2)
            ranges[numTerm][i]['term'] = terms[i]
            graph[terms[i]] = fuzz.trimf(graph.universe, triangle)
        graph.view()
        numTerm += 1
    return ranges
```

```python
print("Функции принадлежности для X:")
X_RANGES = ranges(X_START, X_END, 'Х')
```

```python
print("Функции принадлежности для Y:")
Y_RANGES = ranges(Y_START, Y_END, 'Y')
```

```python
print("Функции принадлежности для Z:")
Z_RANGES = ranges(Z_START, Z_END, 'Z')
```

```python
def rulesPrint(Xterm, Yterm, Zterm, isMamdami, numRange):
    if isMamdami == True:
        print('Если х = {} и y = {} то F= {}'.format(Xterm,
                                                     Xterm,
                                                     Xterm))
    else:
        print('Если х = {} и y = {} то F= {}'.format(Xterm,
                                                     Yterm,
                                                     TERMS[numRange].index(Zterm) - ((numRange * 2 + 4) / 2)))
```

```python
def calculate(Xranges, Yranges, Zranges, isMamdami, numRange):
    for Xrange in Xranges.values():
        for Yrange in Yranges.values():
            x = (Xrange['start'] + Xrange['end']) / 2
            y = (Yrange['start'] + Yrange['end']) / 2
            z = x * math.sin(y)
            for Zrange in Zranges.values():
                if Zrange['start'] < z and z < Zrange['end']:
                    rulesPrint(Xrange['term'], Yrange['term'], Zrange['term'], isMamdami, numRange)
        print("")
```

```python
def system(isMamdami):
    numRange = 0
    for Xranges, Yranges, Zranges in zip(X_RANGES.values(), Y_RANGES.values(), Z_RANGES.values()):
        if isMamdami == True:
            print("------------Мамдани для {} термов------------".format(numRange * 2 + 5))
        else:
            print("------------Cинглтон для {} термов------------".format(numRange* 2 + 5))
        print("")
        calculate(Xranges, Yranges, Zranges, isMamdami, numRange)
        numRange += 1
        print("")
```

```python
system(True)
```
